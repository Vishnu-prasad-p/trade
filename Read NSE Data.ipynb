```python
stocks = NIFTY_50
```

```python
def get_stock(ticker, years = 3):
    start = (datetime.datetime.today() - datetime.timedelta(days=years*365)).date()
    end = datetime.datetime.today().date()
    data = pdr.get_data_yahoo(ticker+".NS", start=start, end=end)
    data[f'{ticker}'] = data["Close"]
    data = data[[f'{ticker}']] 
    return data
```

```python
from functools import reduce

def combine_stocks(tickers):
    data_frames = []
    for i in tickers:
        data_frames.append(get_stock(i))
        
    df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['Date'], how='outer'), data_frames)
    return df_merged
```

```python
portfolio = combine_stocks(stocks)
```

```python
def calculateRSI(portfolio,window=14,symbol = False):
    if symbol:
        portfolio[symbol+'_RSI'] = RSI(portfolio[symbol],window)
    else:
        for symbol in portfolio.columns:
            portfolio[symbol+'_RSI'] = RSI(portfolio[symbol],window)
    return portfolio
```

```python
def RSI(close,window=14):
    delta = close.diff()
    up = delta.clip(lower=0)
    down = -1*delta.clip(upper=0)
    ema_up = up.ewm(com=window-1, adjust=False).mean()
    ema_down = down.ewm(com=window-1, adjust=False).mean()
    rs = ema_up/ema_down
    return 100 - (100/(1 + rs))
```

```python
op = calculateRSI(portfolio,14,'SBIN')
```
